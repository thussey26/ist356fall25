```python
data1 = [
    {
        "name": "Alice",
        "age": 30,
        "address": {
            "street": "123 Main St",
            "city": "New York"
        },
        "hobbies": ["reading", "hiking"]
    },
    {
        "name": "Bob",
        "age": 25,
        "address": {
            "street": "456 Elm St",
            "city": "Los Angeles"
        },
        "hobbies": ["gaming", "cycling"]
    }
]
```

```python
# Q11
# How many rows and columns are in the following data frame? 
# A. 2 rows, 5 columns 

pd.json_normalize(data1)
```

```python
# Q 12
# How many rows and columns are in the following data frame? 

# record_path flattens the nested list of hobbies. Tells pandas to make each hobby its own row.
# meta specifies the other fields to include in each row.
# ['address', 'street'] and ['address', 'city'] tell Pandas to extract nested fields in the address dictionary.
# errors='ignore' means avoids errors if any of the entries donâ€™t contain one of those fields.

df = pd.json_normalize(data1, 
                       record_path='hobbies', 
                       meta=['name', 'age', ['address', 'street'], ['address', 'city']], 
                       errors='ignore')
df
```

```python
# Q 13
# How many rows and columns are in the following data frame? 
# B.	2 rows, 1 column

data = next(item for item in data1 if item['name'] == 'Alice') # list comprehension to find the dictionary where name is Alice
data_df = pd.DataFrame(data['hobbies'], columns=['hobby']) # extract the hobbies list, name column hobby, and create a data frame
data_df
```

```python
# Q23

teststring = 'Find all of the words in a string that are less than 4 letters'
results= []
for word in teststring.split():
    if len(word) < 4:
        results.append(word)
print(results)

# split the string into a list of word
# iterates over each word
# filter to only words less than 4 letters.
# store in a list called results
```

```python
teststring = 'Find all of the words in a string that are less than 4 letters'
results = [word for word in teststring.split() if len(word) < 4]
print(results)
```

### Best practice - data lineage

When combining datasets, its a really good idea to retain **data lineage**, or a record of where the data came from. This can be done by added a column to each dataframe before concatenating.

**KPI** summarizes larger points of data, so they can be measured over time. 
- It is used to measure performance
- Support decision making 
- Track progress over time

In summary a quantifiable way to evaluate progress toward achieving key business or project goals
